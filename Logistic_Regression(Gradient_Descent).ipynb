```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
X,y=make_classification(n_samples=100,n_classes=2,n_clusters_per_class=1,n_features=2,n_informative=1,n_redundant=0,random_state=41,class_sep=30,hypercube=False)
```

# Logistic Regression Using Sigmoid Implementaion

```python
def logistic(X,y):
  X=np.insert(X,0,1 ,axis=1)
  w=np.ones(X[0].shape)
  epochs=1000
  lr=0.1

  def sigmoid(Z):
    return 1/(1+np.exp(-Z))

  for i in range(epochs):
    j=np.random.randint(X.shape[0])
    data=sigmoid(np.dot(X[j],w))
    w=w+lr*(y[j]-data)*X[j]

  return w[0], w[1:]
```

```python
intercept_,coef_=logistic(X,y)
```

```python
m=(-coef_[0]/coef_[1])
b=(-intercept_/coef_[1])
```

```python
X_input=np.linspace(-3,3,100)
y_input=m*X_input+b
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(X_input,y_input)
```

# Logistic Regression Using Gradient Descent


```python
def sigmoid(Z):
  return 1/(1+np.exp(-Z))

def grad(X,y):
  lr=0.5
  epochs=2500
  X=np.insert(X,0,1,axis=1)
  w=np.ones(X[0].shape)



  for i in range(epochs):
    yhat=sigmoid(np.dot(X,w))
    w=w+lr*(np.dot((y-yhat),X)/X.shape[0])

  return w
```

```python
data=grad(X,y)
```

```python
intercept=data[0]
coefficient=data[1:]
```

```python
mval=(-coefficient[0]/coefficient[1])
bval=(-intercept/coefficient[1])
```

```python
X_input_1=np.linspace(-3,3,100)
y_input_1=mval*X_input+bval
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(X_input_1,y_input_1,color='b')
plt.plot(X_input1,y_input1,color='g')
plt.plot(X_input,y_input,color='r')
```
