```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

Logistic Regression from scratch


```python
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_clusters_per_class=1,n_classes=2,class_sep=30,hypercube=False, random_state=41)
```

```python
X.shape[0]
```

```python
def logistic(X,y):
  X=np.insert(X,0,1 ,axis=1)
  w=np.ones(X[0].shape)
  epochs=1000
  lr=0.01
  for i in range(epochs):
    j=np.random.randint(X.shape[0])
    data=sigmoid(np.dot(X[j],w))
    w=w+lr*(y[j]-data)*X[j]
    return w[0],w[1:]

```

```python
def sigmoid(Z):
  return 1/(1+np.exp(-Z))
```

```python
p1,p2=logistic(X,y)
print(p1,p2)
```

```python
m=-p2[0]/p2[1]
c=-p1/p2[1]
```

```python
X_input=np.linspace(-3,3,100)
y_input=m*X_input+c
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=50)
plt.plot(X_input,y_input)
```
